```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
doma = pd.read_csv('doma.csv', sep=';')
```

```python
restaurants = pd.read_excel('Restaurants.xlsx', sheet_name='data')
```

```python
doma.head()
```

```python
restaurants.head()
```

```python
#построю гистограммы

fig = plt.figure(figsize=(10.,8.))
_ = plt.hist(doma["population"], bins = 30, color = "black")
```

```python
fig = plt.figure(figsize=(10.,8.))
_ = plt.hist(restaurants["Y"], bins = 30, color = "black")
```

```python
from numpy import array, linspace
from sklearn.neighbors.kde import KernelDensity
from matplotlib.pyplot import plot

doma1 = np.array(doma["population"]).reshape(-1, 1)
kde = KernelDensity(kernel='cosine', bandwidth=10).fit(doma1)
s = linspace(0,50)
e = kde.score_samples(s.reshape(-1,1))
fig = plt.figure(figsize=(10., 5.))
_ = plot(s, e, color = "black")
```

```python
restaurants1 = np.array(restaurants["Y"]).reshape(-1, 1)
kde = KernelDensity(kernel='tophat', bandwidth=1).fit(restaurants1)
s = linspace(0,50)
e = kde.score_samples(s.reshape(-1,1))
fig = plt.figure(figsize=(10.,5.))
_ = plot(s, e, color = "black")
```

```python
#distplot () используется для визуализации параметрического распределения набора данных
import seaborn as sb
plt.figure(figsize=(10.,5.))
sb.distplot(doma1, kde=True)
plt.show()
```

```python
fig = plt.figure(figsize=(10.,5.))
sb.distplot(restaurants1, kde=True)
plt.show()
```
